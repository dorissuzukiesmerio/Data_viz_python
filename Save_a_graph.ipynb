### Use **ggsave** to save graph as a png file

```r
library(ggplot2)
plot_1 <- ggplot(mtcars, 
            aes(x = wt , y = mpg)) +
  geom_point()
print(plot_1)

# save a graph
ggsave(plot_1, filename = "mygraph_1.png")
```
